Importing necessary libraries

```python
import pandas as pd
import numpy as np
import cv2
import os
```

"Reading" images and save them in pandas dataframe

```python
root_dir = "Train"

# getting the list of folder names (label encodings)
folders = os.listdir(root_dir)

# empty list for the data
train = []

# going over the folders (0, 1, 2, ..., 77)
for sub_folder in folders[39:]:

  # getting the list of image names in each folder
  images = os.listdir(os.path.join(root_dir, sub_folder))

  # going over the images
  for image_name in images:
    # reading the image
    image = cv2.imread(os.path.join(root_dir, sub_folder, image_name), 0)

    # we also need to make a feature vector out of the image matrix
    # so that we convert the 64x64 matrix into a vector of size 4096
    # Hint: you can look for the flatten or reshape methods from numpy
    feature_vector = image.flatten()

    # now we need to add the resulting feature vector and
    # its label (here its the sub_folder name) into the train list
    # note that '*' is used for unpacking
    train.append([*feature_vector, int(sub_folder)])

train = pd.DataFrame(train)
```

Shuffle data

```python
train = train.sample(frac=1)
```

Save the data in csv file for further work.

```python
train.to_csv('data.csv', index=False)
```
